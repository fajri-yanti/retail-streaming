```python
import uuid
from confluent_kafka import Producer
from faker import Faker
import json
from time import sleep

faker = Faker()
```

```python
class DataGenerator:
    @staticmethod
    def get_data_record():
        record = {
            "order_date": faker.date_this_year().isoformat(),
            "order_id": str(uuid.uuid4()),
            "customer_name": faker.name(),
            "product": faker.random_element(elements=('Laptop', 'Smartphone', 'Tablet')),
            "category": faker.random_element(elements=('Electronics', 'Accessories')),
            "price_per_product": faker.random_int(min=50000, max=2000000),
            "total_payment": faker.random_int(min=1, max=10) * faker.random_int(min=50000, max=2000000),
            "payment_type": faker.random_element(elements=('Credit Card', 'Cash', 'E-Wallet')),
            "timestamp": faker.unix_time(),
            "store_branch": faker.random_element(elements=('Branch A', 'Branch B', 'Branch C')),
            "is_new": True
        }
        return record

producer = Producer({'bootstrap.servers': 'kafka_broker_address:9092'})
kafka_topic_partition = 'transaction_group'

for _ in range(400):
    data_record = DataGenerator.get_data_record()
    producer.produce(topic=kafka_topic_partition, value=json.dumps(data_record).encode('utf-8'))
    print(f"Produced record with Order ID: {data_record['order_id']}, Product: {data_record['product']}, Category: {data_record['category']}, Order Date: {data_record['order_date']}")
    sleep(5)

producer.flush()
```
